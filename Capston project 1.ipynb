```python
df_clean = df.drop(columns=[
    "Reason for cancelling by Customer",
    "Driver Cancellation Reason",
    "Incomplete Rides Reason"
])
```

```python
df_clean.isnull().sum()
```

```python
df_clean["Booking Value"].fillna(df_clean["Booking Value"].median(), inplace=True)
df_clean["Ride Distance"].fillna(df_clean["Ride Distance"].median(), inplace=True)
df_clean["Payment Method"].fillna(df_clean["Payment Method"].mode()[0], inplace=True)
```

```python
df_clean["Driver Ratings"].fillna(df_clean["Driver Ratings"].median(), inplace=True)
df_clean["Customer Rating"].fillna(df_clean["Customer Rating"].median(), inplace=True)
```

```python
df_clean["Incomplete Rides"].fillna(df_clean["Incomplete Rides"].mode()[0], inplace=True)
```

```python
df_clean["Avg CTAT"].fillna(df_clean["Avg CTAT"].median(), inplace=True)
df_clean["Avg VTAT"].fillna(df_clean["Avg VTAT"].median(), inplace=True)
```

```python
df_clean["Cancelled Rides by Driver"].fillna(df_clean["Cancelled Rides by Driver"].mode()[0], inplace=True)
df_clean["Cancelled Rides by Customer"].fillna(df_clean["Cancelled Rides by Customer"].mode()[0], inplace=True)
```
